```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from keras.datasets import cifar10

(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.99, random_state=123)
y_train.shape
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
y_train = y_train.ravel()

rf_classifier = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, random_state=123)


rf_classifier.fit(X_train, y_train)
importances = rf_classifier.feature_importances_
importance_image = importances.reshape(32, 32, 3)

normalized_importance_image = (importance_image - np.min(importance_image)) / (np.max(importance_image) - np.min(importance_image))
plt.imshow(normalized_importance_image)
plt.title("Feature Importances for CIFAR-10")
plt.axis("off")
plt.show()
```
